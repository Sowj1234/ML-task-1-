```python
taxi=p.drop("num_of_passengers", axis=1)
taxi_labels=p["num_of_passengers"].copy()
```

```python
taxi.describe() 
```

```python
median=taxi['distance_traveled'].median()
```

```python
taxi['distance_traveled'].fillna(median)
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='median')
imputer.fit(taxi)
```

```python
imputer.statistics_.shape
```

```python
X=imputer.transform(taxi)
```

```python
taxi_tr = pd.DataFrame(X, columns=taxi.columns)
```

```python
taxi_tr.info()
```

```python
taxi_tr.describe()
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
my_pipeline =Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler())
])
```

```python
taxi_num_tr=my_pipeline.fit_transform(taxi_tr)
```

```python
taxi_num_tr
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model=DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(taxi_num_tr,taxi_labels)
```

```python
some_data=taxi.iloc[:5]
```

```python
some_labels=taxi_labels.iloc[:5]
```

```python
prepared_data=my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
taxi_predictions = model.predict(taxi_num_tr)
mse= mean_squared_error(taxi_labels,taxi_predictions)
rmse= np.sqrt(mse)
```

```python
rmse
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,taxi_num_tr,taxi_labels,scoring="neg_mean_squared_error")
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores 
```

```python
def print_scores(scores):
    print("scores : " , scores)
    print("mean : ", scores.mean())
    print("Standard Deviation :", scores.std())
```

```python
print_scores(rmse_scores)
```
